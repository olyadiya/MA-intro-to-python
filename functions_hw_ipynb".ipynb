````python
# здесь пишем функцию preprocess_text
# функция принимает на вход текст и возвращает очищенный текст
# убедитесь в том, что функция выполняет всё по требованиям:
# 1. удаляет HTML-теги
# 2. удаляет специальные символы
#    (кроме цифр, дефисов, точек, запятых, восклицательных, вопросительных знаков, двоеточий, точек с запятой, многоточий)
# 3. приводит текст к нижнему регистру
# 4. заменяет множественные пробелы на один
def preprocess_text(shakespeares_sonnets_TXT_FolgerShakespeare):
    text = shakespeares_sonnets_TXT_FolgerShakespeare
    if not isinstance(text, str):
        return ""

    text = re.sub(r'<[^>]+>', ' ', text)
    text = re.sub(r'[^a-zA-Zа-яА-ЯёЁ0-9\s\-.,!?:;]', ' ', text)
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()

    return text
````

### **4: частотный словарь**

Перед вами стандартный образец создания частотного словаря:

```python
from collections import Counter

freq_dict = Counter(words)
```

В этом примере в переменной words хранится список слов текста

Counter автоматически считает, сколько раз каждое слово встречается в этом списке, и возвращает словарь вида

```
{
    word: count
}
```

\- где word - слово

\- count - его частотность в списке words

Напишите функцию, которая применяет Counter к вашему очищенному списку слов

Запустите её

````python
# напишите функцию get_word_frequencies,
# которая принимает на выход текст
# делит его на слова и возвращает частотный словарь

from collections import Counter

def get_word_frequencies(text):
    words = text.split()
    freq_dict = Counter(words)
    return dict(freq_dict)

data = load_dataset("shakespeares_sonnets_TXT_FolgerShakespeare.txt")
cleaned_text = preprocess_text(data)
frequencies = get_word_frequencies(cleaned_text)
for word, count in sorted(frequencies.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"  '{word}': {count}")
````

````python
import re
from collections import Counter

def load_data(shakespeares_sonnets_TXT_FolgerShakespeare):

    with open(shakespeares_sonnets_TXT_FolgerShakespeare, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

def clean_text(text):

    text = re.sub(r'<.*?>', '', text)
    return text

def tokenize_text(text):
    text = text.lower()
    # Оставляет только буквы, знаки препинания и пробелы
    text = re.sub(r'[^а-яёa-z0-9.,!?;: -]', ' ', text)
    # Удаление повторяющихся пробелов
    text = re.sub(r'\s+', ' ', text).strip()
    # Приведение к нижнему регистру

    return text

def create_frequency_dict(text):

    cleaned = re.sub(r'[^\w\s]', ' ', text)
    words = cleaned.split()
    freq_dict = Counter(words)

    return freq_dict

text = load_data("shakespeares_sonnets_TXT_FolgerShakespeare.txt")
print(f"Длина файла: {len(text)}")
print("Первые 10 символов:", text[:50])

tokens = tokenize_text(text)
print("После токенизации:", tokens[:200])

cleaned_text = clean_text(text)
freq_dict = create_frequency_dict(tokens)
print("Самые частые слова:", freq_dict.most_common(5))

````
