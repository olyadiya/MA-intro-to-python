```python
numbers = []
while True:
    user_input = input("Введите числа2: ")
    input_list = user_input.split()
    filtered_numbers = []
    for item in input_list:
        try:
            num = int(item)
            if num >= 0:
                filtered_numbers.append(num)
        except ValueError:
            continue
    numbers.extend(filtered_numbers)
    if '0' in input_list:
        break
print(f"список чисел: {numbers}")
```

```python
celsius = [0, 15, 25, 30, 40, 100]

fahrenheit = [(c * 9/5 + 32) for c in celsius]

print(f"Температура в Цельсиях: {celsius}")
print(f"Температура в Фаренгейтах: {fahrenheit}")
```

```python
grades = [45, 85, 92, 33, 67, 78, 90, 55, 29, 88]
passed = [grade for grade in grades if grade >= 60]
print(f"Сданные экзамены: {passed}")
```

```python
transactions = [100, -50, 200, -30, 150, -20, 300]
taxes = [amount * 0.15 for amount in transactions if amount > 0]
print(f"Налоги: {taxes}")
```

```python
products = ['футболка', 'кружка', 'блокнот']
colors = ['красный', 'синий', 'зеленый']
combinations = [f"{product} {color}" for product in products for color in colors]

print("Все комбинации товаров и цветов:")
for combo in combinations:
    print(combo)
```

```python
employees = [('Иван', 45), ('Мария', 92), ('Петр', 33), ('Анна', 67)]
performance = [
    "Отлично" if percent >= 90
    else "Хорошо" if percent >= 60
    else "Требует улучшения"
    for _, percent in employees
]
for i, (name, percent) in enumerate(employees):
    print(f"{name}: {percent}% - {performance[i]}")
```

```python
sales = [
    {'name': 'Алексей', 'sales': 150, 'returns': 10},
    {'name': 'Ольга', 'sales': 200, 'returns': 5},
    {'name': 'Дмитрий', 'sales': 80, 'returns': 25},
    {'name': 'Елена', 'sales': 300, 'returns': 8}
]

top_of_the_top = [
    manager['name']
    for manager in sales
    if (manager['sales'] - manager['returns']) >= 150
    and (manager['returns'] / manager['sales'] * 100) < 10
]
for manager in top_of_the_top:
    print(f"Топ-менеджер {manager}")
```
