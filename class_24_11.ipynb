```python
shopping_list = []
command = ''
while command != '4':
  command = input("Введите номер команды: 1. Добавить товар; 2. Удалить товар; 3. Показать список; 4. Выйти. ")
  if command == '1':
    item = input()
    shopping_list.append(item)

  elif command == '2':
    item = input()
    if item in shopping_list:
      shopping_list.remove(item)
    else:
      print("такого товара нет в списке")

  elif command == '3':
    print(shopping_list)

  elif command == '4':
    print('вы вышли из программы')


```
